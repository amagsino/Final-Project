```python
import pandas as pd
```

```python
#Reading tobacco csv
tobacco = pd.read_csv('original_datasets/Behavioral_Risk_Factor_Data__Tobacco_Use__2011_to_present_.csv')
tobacco.head()
```

```python
#Filter out the data to get correct dataset
tobacco_df = tobacco[(tobacco.TopicDesc == 'Cigarette Use (Adults)') & (tobacco.MeasureDesc == 'Current Smoking')]
tobacco_df.head()
```

```python
#Filter out the data to get correct dataset
tobacco_df1 = tobacco_df[(tobacco_df.Gender == 'Overall') & (tobacco_df.Race == 'All Races') & (tobacco_df.Age == 'All Ages')]
tobacco_df1.head()
```

```python
#selecting the columns to use
tobacco_df2 = tobacco_df1[["YEAR", "LocationDesc", "Data_Value"]]
tobacco_df2.head()
```

```python
#re-name columns
tobacco_df2.columns = ["year", "state", "tobacco_percentage"]
tobacco_df2.head()
```

```python
#Drop NA
tobacco_clean=tobacco_df2.dropna()
tobacco_clean
```

```python
#Sort dataframe and re-arrange the column names
tobacco_clean1 = tobacco_clean.sort_values(['state', 'year'])
tobacco_final = tobacco_clean1[['state','year', 'tobacco_percentage']]
tobacco_final['year']= pd.to_datetime(tobacco_final['year'], format='%Y').dt.year
tobacco_final
```

```python
#reading csv for Per Capita income
income = pd.read_csv('original_datasets/per_cap_income_us.csv')
income.head()
```

```python
#Droping NA
per_cap_df_dropped=income.dropna()
per_cap_df_dropped.head()
```

```python
#Drop GeoFips Column
drop_col=per_cap_df_dropped.drop('GeoFips', axis=1)
drop_col.head()
```

```python
#re-arrange the columns for the merge
drop_col.columns=['state', 'year', 'per capita income']
drop_col.head()
```

```python
#Remove asterik
remove_ast=drop_col.replace('\*','',regex=True).astype(str)
remove_ast['state'] = remove_ast['state'].str.strip()
remove_ast.head()
```

```python
#remove decimals
remove_ast['year'] = remove_ast['year'].astype(str).replace('\.0', '', regex=True)
remove_ast.head()
```

```python
#Rename dataframe and create year as an integer
cleaned_per_cap_income=remove_ast
cleaned_per_cap_income.year.astype(int)
cleaned_per_cap_income.sort_values(by=['year'])
cleaned_per_cap_income['year']= pd.to_datetime(cleaned_per_cap_income['year'], format='%Y').dt.year
cleaned_per_cap_income
```

```python
#Read csv for deaths data
deaths = pd.read_csv('original_datasets/leading_causes_death_us.csv')
deaths.head()
```

```python
#Select the columns
deaths_df = deaths[["Year", "State", "Deaths"]]
deaths_df.head()
```

```python
#rename columns
deaths_df.columns = ["year", "state", "Deaths"]
deaths_df.head()
```

```python
#re-arrange column order for the merge and convert year column to a datetime dtype
deaths_df1 = deaths_df[['state','year', 'Deaths']]
deaths_df1['year']= pd.to_datetime(deaths_df1['year'], format='%Y').dt.year
deaths_df1
```

```python
#Drop NA
deaths_clean=deaths_df1.dropna()
deaths_clean
```

```python
#Read csv for clean obesity data since calculations where done to get values
obesity= pd.read_csv('clean_datasets/clean_obesity.csv')
# obesity.year.astype(int)
obesity['year']= pd.to_datetime(obesity['year'], format='%Y').dt.year 

obesity
```

```python
#Merging Obesity dataframe and tobacco
all_data_st1 = pd.merge(obesity, tobacco_final, on=['state', 'year'])

all_data_st1
```

```python
#Merging per Capita income
all_data_st2 = pd.merge(all_data_st1, cleaned_per_cap_income, on=['state', 'year'], how='left')
all_data_st2
```

```python
#merging Deaths dataframe
final_df = pd.merge(all_data_st2, deaths_clean, on=['state', 'year'], how='left')
final_df
```

```python
#writting final csv 
final_df.to_csv(r'clean_datasets/merged_results.csv', index = False)
final_df
```
